```python
file = open(r'E:\Users\user\Jupyter\練習資料\words.txt', 'r')
wordcount={}
word_list=file.read().split()
for word in word_list:
    if word not in wordcount:
        wordcount[word] = 1
    else:
        wordcount[word] += 1     
        
for k,v in wordcount.items():
    print(k,v)
```
